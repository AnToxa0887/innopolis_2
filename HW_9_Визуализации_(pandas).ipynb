Файл df3.csv находится в архиве с заданием

```python
# Запустите эту ячейку
import pandas as pd
%matplotlib inline

df3 = pd.read_csv('df3.csv')
print(len(df3))
print(df3.head())
```

Эти данные представляют собой заводские производственные показатели и сообщенное количество дефектов в определенные дни недели.

### 1. Воссоздайте этот график разброса «произведенных» и «дефектных»  деталей. Обратите внимание на цвет и размер точек. Также обратите внимание на размер картинки. Сможете ли растянуть его аналогичным образом?

```python
# ВАШ КОД
import matplotlib.pyplot as plt
plt.figure(figsize=(12,3))
plt.scatter(df3.produced, df3.defective, c = 'red', s=20)
plt.xlabel("produced")
plt.ylabel("defective")
```

```python
# ОЖИДАЕМЫЙ РЕЗУЛЬТАТ
```

### 2. Создайте гистограмму для колонки 'produced'

```python
# ВАШ КОД
plt.hist(df3.produced)
plt.ylabel("Frequency")
```

```python
# ОЖИДАЕМЫЙ РЕЗУЛЬТАТ
```

### 3. Перерисуйте гистограмму как на рисунке ниже

```python
# ВАШ КОД
plt.hist(df3.produced, edgecolor='black', range=[55, 95]) #задаем цвет контура бара с помощью параметра 'edgecolor'и min/max точки гистограммы
plt.ylabel("Frequency") #название оси Y
plt.xlim([55, 95]) #ограничение оси X
```

```python
# ОЖИДАЕМЫЙ РЕЗУЛЬТАТ
```
